```python
import advertools as adv
import requests
from bs4 import BeautifulSoup
import pandas as pd
from csv import DictWriter
import csv

```

Definim Variabilele: sitemap_url adresa url catre sitemap index

```python
url = "https://www.drool.ro/sitemap.xml"
```

```python
# PT SCRAPING SAFE
headers = {
    "User-Agent": "Firefox/12.0",
    "HTTP headers Accept": "text/html",
    "HTTP header Referer": "http://www.google.com/",
    "Accept-Language": "en-gb"
}
```

Scoatem Adresele URL din Sitemap si Facem export in CSV 



```python
sitemap = adv.sitemap_to_df(url)
sitemap.to_csv("sitemap.csv")
sitemap_df = pd.read_csv("sitemap.csv", index_col=False)
sitemap_df.drop(columns=["Unnamed: 0"], inplace=True)
sitemap_df
```

```python
sitemapSoup = BeautifulSoup(requests.get(url, headers=headers).content, 'lxml')
sitemapurls = sitemapSoup.find_all('loc')
xml_urls = [myurl.text for myurl in sitemapurls]

url_list = sitemap_df['loc']
```

```python
headersCSV = ['ID', 'Adresa URL','Response Code', 'Canonical',"Self Canonical"]
```

```python
def write_to_output_csv(row):
    with open('output.csv', 'a', newline='') as f_object:
        # Pass the CSV  file object to the Dictwriter() function
        # Result - a DictWriter object
        dictwriter_object = DictWriter(f_object, fieldnames=headersCSV)
        # Pass the data in the dictionary as an argument into the writerow() function
        dictwriter_object.writerow(row)
        # Close the file object
        f_object.close()
```

```python
for i in range(len(url_list)):
  if (type(url_list[i])) != str:
    del url_list[i]
```

```python
id = 1
write_to_output_csv(row={"ID":"ID", 'Adresa URL':'Adresa URL','Response Code':'Response Code','Canonical':'Canonical',"Self Canonical":"Self Canonical"})
```

```python
for entry in range(id,len(url_list)):
  websiteurls = url_list[entry]
  x = requests.get(websiteurls, headers=headers)
  source = BeautifulSoup(x.text, 'html.parser')
  r = {
      "ID": id,
      "Adresa URL": websiteurls,
      "Response Code":x.status_code,
      "Canonical": source.find('link', {'rel': 'canonical'})['href'],
      "Self Canonical": websiteurls == source.find('link', {'rel': 'canonical'})['href']
      }
  write_to_output_csv(r)
  id +=1
  print(id, "/ ", len(url_list))
```
